## Multi-index & Index hierarchy

```python
outside = ["G1", "G1", "G1", "G2", "G2", "G2"]
inside = [1, 2, 3, 1, 2, 3]
hier_idx = list(zip(outside, inside))
hier_idx = pd.MultiIndex.from_tuples(hier_idx)
```

```python
rng = np.random.default_rng(seed=101)
data = rng.normal(size=(6,2))

hier_df = pd.DataFrame(
    data=data,
    index=hier_idx,
    columns=["A", "B"]
)
hier_df
```

```python
hier_df.loc["G1"]
```

```python
hier_df.loc["G1"].loc[1]
```

```python
hier_df.index.names
```

```python
hier_df.index.names = ["Groups", "Num"]
hier_df
```

```python
hier_df.loc["G2"].loc[2, "B"]
```

```python
# xs : cross-section function for multi-index dataframes
hier_df.xs("G1")
```

```python
hier_df.xs(1, level="Num")
```

# Groupby

```python
data = {
     "Company": ["GOOG", "GOOG", "MSFT", "MSFT", "FB", "FB"],
     "Person": ["Sam", "Charlie", "Amy", "Vanessa", "Carl", "Sarah"],
     "Sales": [200, 120, 340, 124, 243, 350]
 }
```

```python
df = pd.DataFrame(data)
df
```

```python
df.groupby(by="Company").mean()
```

```python
# ATTENTION : counts objects as well as numbers
df.groupby(by="Company").count()
```

```python
# ATTENTION : strings are returned by order
df.groupby(by="Company").max()
```

```python
df.groupby(by="Company").describe()
```

# Merge, join, concatenate

```python
data = {
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
    }

index = [0, 1, 2, 3]

df1 = pd.DataFrame(data, index)
df1
```

```python
data = {
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
    }

index = [4, 5, 6, 7]

df2 = pd.DataFrame(data, index)
df2 
```

```python
data = {
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'D': ['D8', 'D9', 'D10', 'D11']
    }

index=[8, 9, 10, 11]

df3 = pd.DataFrame(data, index)
df3
```

```python
arr_df = [df1, df2, df3]
pd.concat(arr_df)
```

```python
pd.concat(arr_df, axis=1)
```

MERGE : SQL style join | join done on columns

   - inner : intersection of keys

   - outer : union of keys

   - left : use keys from left frame only
   
   - right : use keys from right frame only

```python
d = {
    "key": ["K0", "K1", "K2", "K3"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]
}

left = pd.DataFrame(d)
left
```

```python
d = {
    "key": ["K0", "K1", "K2", "K3"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"]
}
right = pd.DataFrame(d)
right
```

```python
pd.merge(left, right, how="inner", on="key")
```

```python
d = {
    "key1": ["K0", "K0", "K1", "K2"],
    "key2": ["K0", "K1", "K0", "K1"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]
}

left = pd.DataFrame(d)
left
```

```python
d = {
    "key1": ["K0", "K1", "K1", "K2"],
    "key2": ["K0", "K0", "K0", "K0"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"]
}

right = pd.DataFrame(d)
right
```

```python
pd.merge(left, right, how="inner", on=["key1", "key2"])
```

```python
pd.merge(left, right, how="outer", on=["key1", "key2"])
```

```python
pd.merge(left, right, how="left", on=["key1", "key2"])
```

```python
pd.merge(left, right, how="right", on=["key1", "key2"])
```

JOIN : same as merge, except join done on indexes

```python
d = {
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"]
}
index = ["K0", "K1", "K2"]

left = pd.DataFrame(d, index)
left
```

```python
d = {
    "C": ["C0", "C2", "C3"],
    "D": ["D0", "D2", "D3"]
}
index = ["K0", "K2", "K3"]

right = pd.DataFrame(d, index)
right
```

```python
left.join(right, how="left")
```

```python
left.join(right, how="outer")
```

## Pivot table

```python
d = {
    "A": ["foo", "foo", "foo", "bar", "bar", "bar"],
    "B": ["one", "one", "two", "two", "one", "one"],
    "C": ["x", "y", "x", "y", "x", "y"],
    "D": [1, 3, 2, 5, 4, 1]
}

df = pd.DataFrame(d)
df
```

```python
df.pivot_table(values="D", index=["A", "B"], columns="C")
```

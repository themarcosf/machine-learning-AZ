# Dataviz

One of the main tasks of a data scientist is to present the results and conclusions of the analysis in an intuitive and visual way. For this reason, it is very important that **data visualization** techniques are applied. This is an area of data science in itself, often called **dataviz**, and is of enormous importance for a data science project to be successful and **well communicated**.

Further, the ability to visualize the data helps a lot in the **EDA** process, after all, visual representations of the data are always simpler and more intuitive to interpret!

So, the main objectives of dataviz are:

- support the exploratory data analysis (EDA)
- Help in the presentation of results and conclusions

## Matplotlib

Matplotlib is the most popular library for data visualization in Python. It allows the generation of different types of graphs, such as line, bar, pie, scatter and histogram graphs. It also allows personalizing the graphs in many ways, adding **multiple data sources**, **changing color**, **marker format** or generating **multiple graphs in the same image**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
x_points = [1, 2, 3, 4, 5]
y_points = [1, 4, 9, 16, 25]
plt.plot(x_points, y_points, marker="o");
```

```python
x_points = np.linspace(-100, 100, 400)
y_points = x_points ** 2
plt.plot(x_points, y_points, linestyle="--", color="green", linewidth=0.5);
```

The definition of the values on the x axis is optional. If you do not specify the values of x, matplotlib takes the **indices of the array for the x axis** and the **values for the y axis**

```python
array = np.linspace(0, 10, 100)
plt.plot(array);
```

Matplotlib allows us to **change the size of the graph** and also **plot more than one curve on the same axis**

```python
figure = plt.figure(figsize=(15,8))
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, 5*np.sin(x));
plt.plot(x, np.cos(x));
```

It is also possible to add titles and legends to the graph and axes.

```python
plt.figure(figsize=(8,4))
plt.title("Sine and cosine functions")

plt.plot(x, np.sin(x), label = "f(x) = sin(x)", color="red")
plt.plot(x, np.cos(x), label = "f(x) = cos(x)")

plt.legend()

plt.xlabel("x")
plt.ylabel("y = f(x)")

plt.show();
```

Sometimes we are not interested in displaying all the data in the same graph. However, we would like the different graphs to be close together to be viewed simultaneously and compared.

With matplotlib, it is possible to make several different graphs in blocks. For this, we need to define a **grid**, and different **axes** to make the plots, as we see below.

```python
gridsize = (2, 2)

# the arguments to plt.subplot2grid are:
# - the size of the grid (a tuple)
# - the position of the axis in the grid (a tuple (row, column))
# - optional: colspan, rowspan

ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2, rowspan=1)
ax2 = plt.subplot2grid(gridsize, (1, 0))
ax3 = plt.subplot2grid(gridsize, (1, 1))

p1, = ax1.plot(x, np.sin(x), color="red", label="sin(x)")
p2, = ax2.plot(x, np.cos(x), label = "cos(x)")
p3, = ax3.plot(x, x**2, color="green", label="x^2")

plt.legend((p1, p2, p3), (p1.get_label(), p2.get_label(), p3.get_label()), 
           bbox_to_anchor=(1, 1), loc="center left")

plt.tight_layout()
plt.show();
```

Another way to make multiple graphs in the same figure is using **plt.subplots()**. This function has a parameter with a 3-digit number: the first says how many lines we will have, the second says how many columns, and the third says which graph we are in. The graph counter starts at 1, with the first graph being the upper left corner. The index grows from left to right, and from top to bottom. To edit each graph individually, just call `plt.subplot` passing the index of the desired graph. From that point on, all operations will be on that graph. See the example below, which plots 4 graphs:

```python
# 2 lines, 1 column, we are in the 1st graph
plt.subplot(211)
plt.plot(x, np.sin(x), color="red", label="sin(x)")
plt.title('Sine')
plt.xlabel('x')
plt.ylabel('y = sin(x)')

# 2 lines, 2 columns, we are in the 3rd graph (because the first graph occupied two cells)
plt.subplot(223)
plt.plot(x, np.cos(x), label = "cos(x)")
plt.title('Cosine')
plt.xlabel('x')
plt.ylabel('y = cos(x)')

# 2 lines, 2 columns, we are in the 4th graph (because the first graph occupied two cells)
plt.subplot(224)
plt.plot(x, x**2, color="green", label="x^2")
plt.title('Parabole')
plt.xlabel('x')
plt.ylabel('y = x^2')

plt.tight_layout()
plt.show()
```

It is also possible to make **bar graphs**:

```python
courses = ['Python', 'Data Science', 'Web Full Stack', 'Database', 'Agile']

enrolled = [122, 47, 39, 14, 19]

plt.bar(courses, enrolled)
plt.title('Enrolled in programming courses')
plt.xlabel('Course')
plt.ylabel('Number of enrolled students')

plt.xticks(rotation=15)

plt.tight_layout()
plt.show();
```

And also **pie charts**.

```python
courses = ['Python', 'Data Science', 'Web Full Stack', 'Banco de Dados', 'Metodologias geis']
enrolled = [122, 47, 39, 14, 19]

plt.pie(enrolled, labels = courses, autopct='%.2f%%')
plt.title('Enrolled in programming courses')
plt.axis('equal')
plt.tight_layout()
plt.show();
```

```python
courses = ['Python', 'Data Science', 'Web Full Stack', 'Banco de Dados', 'Metodologias geis']
enrolled = [122, 47, 39, 14, 19]

plt.pie(enrolled, autopct=lambda x: '{:.0f}'.format(x*sum(enrolled)/100))
plt.title('Programming courses enrollments')
plt.axis('equal')
plt.legend(courses, bbox_to_anchor=(1,0.5), loc="center left")
plt.tight_layout()
plt.show();
```

## Distribution plots

As the name suggests, they are graphs that help us visualize the **distribution of data**, that is, how the data is distributed in the different values that certain variables can assume.


```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

__Histplot (antigo "distplot")__

Este tipo de grfico ilustra a **distribuio dos dados numricos**, atravs de um histograma.
This type of graph illustrates the **distribution of numerical data**, through a histogram.

As a default, the graph no longer automatically produces the [KDE](https://en.wikipedia.org/wiki/Kernel_density_estimation) curve of the distribution, that is, an **estimate for the probability density function of the distribution.**

```python
tips['total_bill'].describe()
```

```python
sns.histplot(tips['total_bill'], kde=True);
```

One important parameter is "bins", which indicates the number of bars to be included in the histogram. The default parameter tries to find a suitable number of bins, such that it smooths the distribution.

```python
sns.histplot(tips['total_bill'], bins=10, kde=True);
```

Let's see the distribution of other numerical columns.

```python
for col in tips.columns:
    if tips[col].dtype in [np.float64, np.int64]:
        sns.histplot(tips[col], kde=True)
        plt.title("Data distribution in " + col)
        plt.show();
```

__Jointplot__

Jointplot is a graph that relates two different columns. In each of the axes, the distributions of the respective columns are displayed, and between the axes some type of graph is shown for the pairs of points of the two columns. 

This graph in the middle is defined by the "kind" parameter, which can take 6 different values.

Through this graph we can quickly examine the **relationship between the different variables**. This way, we can already know what to expect from the subsequent analyzes.

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.show()

sns.jointplot(x='total_bill', y='tip', data=tips, kind='hist')
plt.show()

sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()

sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()

sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
plt.show()

sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
plt.show()
```

One very interesting argument is "hue", which **separates the data** according to some **categorical** variable. This functionality helps us to clearly visualize the relationship between **numerical** variables with **categorical** variables.

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue="sex");
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue="time");
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue="smoker");
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue="day");
```

```python
tips["weekday"] = tips["day"].apply(lambda x: "Weekend" if x in ['Sun', 'Sat'] else "Workday")
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue="weekday");
```

__Pairplot__

The pairplot produces a matrix of graphs, where the main diagonal will have the **distribution** of the data of the respective column, and the other elements are **relationship graphs** between the different columns. Both graphs can be controlled by the arguments "diag_kind" and "kind", respectively

```python
sns.pairplot(tips, kind="reg", diag_kind="kde");
```

```python
sns.pairplot(tips, hue='sex');
```

## Cathegorical plots

__Countplot__

This graph is a histogram of categorical variables. It is very useful to see the **distribution of the data** in the different categories.

```python
display(tips["sex"].value_counts())
sns.countplot(x=tips["sex"]);
```

```python
display(tips["day"].value_counts())
sns.countplot(data = tips, x="day", hue="sex", order=["Sat", "Sun", "Thur", "Fri"]);
```

__Barplot__

This is the bar graph, similar to the countplot, but with the difference that on the x axis we will have categorical data, and on the y axis some numerical data. What the barplot shows is **some descriptive statistic** (mean, median, etc...) of the numerical variable, separated by the categorical variable (as if it were a "hue").

```python
sns.barplot(x=tips['sex'], y=tips['total_bill']);
```

It is possible to change the descriptive function:

```python
sns.barplot(x=tips['sex'], y=tips['total_bill'], estimator=np.median);
```

__Boxplot__

The "boxplots" graphs illustrate the **distribution of numerical data between the different values of categorical data**. In these graphs, it is possible to visualize the **quartiles**, as well as the **outliers**.

<img src="https://miro.medium.com/max/18000/1*2c21SkzJMf3frPXPAR_gZA.png" width=500>

```python
sns.boxplot(data=tips, x = "sex", y = "total_bill");
```

```python
plt.figure(figsize=(9, 5))
sns.boxplot(x = tips["day"], y = tips["total_bill"], hue=tips["sex"]);
```

```python
plt.figure(figsize=(6, 7))
sns.boxplot(y = tips["day"], x = tips["total_bill"], hue=tips["sex"], orient="h")
```

__Violinplot__

This type of graph is very similar to the boxplot in its use: it illustrates **the distribution of numerical data between the different classes of categorical data**, but explicitly, displaying the KDE curve of the distribution:

```python
sns.violinplot(data = tips, x = "day", y = "total_bill");
```

```python
sns.violinplot(data = tips, x = "day", y = "total_bill", hue = "sex");
```

The "split" argument, when True, shows the distribution of each hue in the same "violin":

```python
sns.violinplot(x = tips["day"], y = tips["total_bill"], hue=tips["sex"], split=True);
```

```python
tips[(tips["day"] == "Sun")]["time"].value_counts()
```

```python
sns.violinplot(x = tips["day"], y = tips["total_bill"], hue=tips["time"], split=True, 
               palette='winter');
```

## Matrix plots

We use matrix graphs to pictorially visualize the relationship between numbers in a matrix, on a color scale. The main type of graph for this purpose is the **heatmap**.

__Heatmap__

This type of graph creates a __color scale__ in which the smallest value will correspond to the lightest color, and the largest value, to the darkest color, of a matrix.

Let's consider, for example, the **correlation between the numerical variables** of the base. We'll see this later in the course, but the correction is a number that goes from -1 to 1, and indicates how much the variables are related, positively or negatively.

To do this, just use the `.corr()` method:

```python
sns.heatmap(tips.corr(), annot=True);
```

```python
sns.heatmap(tips.corr(), vmin=-1, vmax=1);
```
